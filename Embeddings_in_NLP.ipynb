### ***-> Stemming*** --> **Porter Stemming Algorithm**

--> Processing of identifying main word stem after stripping of its affixes, to have meaning closer to its original forms.

```python
from nltk.stem.porter import PorterStemmer
```

```python
stemmer = PorterStemmer()

# define list of words for stemming
words = ['caresses', 'flies', 'dies', 'mules', 'denied','died', 'agreed', 'owned', 
         'humbled', 'sized', 'meetings', 'stating', 'siezing', 'itemization',
         'sensational', 'traditional', 'reference', 'colonizer', 'plotted'
         ]
```

```python
stemmed_words = [stemmer.stem(word) for word in words]
stemmed_words
```

### ***-> Lemmatization***

--> Processing of finding original form a word that you find as head word in a dictionary. It creates the base form of the word before inflection.

```python
from nltk.stem import WordNetLemmatizer

# downlaod the lemmatizer
nltk.download("wordnet")
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
lemmatized_words
```
